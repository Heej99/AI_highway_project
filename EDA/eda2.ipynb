## 2. 파일 이름 앞에 해당 지사 붙여주기

```python
def findfilename(path):
    for filename in os.listdir(path):
        word = filename[:-6] # _S 없애기
        print(word)
        temp_name=ai_df[ai_df['탐지ID']==word].iloc[0,5]
        new_name = temp_name+"_"+word+'.jpg'
        os.rename(path+filename,path+new_name)
```
